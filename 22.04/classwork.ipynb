```python
def to_trinumeric(n):
    answer = ""
    while n:
        answer += str(n % 3)
        n //= 3
    return answer[::-1]


def build_num(n):
    raw_n = to_trinumeric(n)
    if not n % 3:
        raw_n += raw_n[-2:]
    else:
        raw_n += to_trinumeric((n % 3) * 3)

    return int(raw_n, 3)


for n in range(1, 10_000):
    if build_num(n) <= 150:
        result = n

print(result)
# Answer: 16
```

```python
for n in range(4, 10_000):
    string = "4" + "2" * n

    while "42" in string or "8222" in string or "2222" in string:
        if "42" in string:
            string = string.replace("42", "2", 1)
        if "8222" in string:
            string = string.replace("8222", "24", 1)
        if "2222" in string:
            string = string.replace("2222", "8", 1)

    if sum(map(int, string)) == 110:
        print(n)
        break
# Answer: 1591 
```

```python
with open("files/17.txt") as file:
    seq = list(map(int, file))

sub_max = min(elem for elem in seq if elem > 0 and 1000 <= elem < 10_000 and elem % 10 == 6)
amount = max_sum = 0
for i in range(len(seq) - 2):
    if all((
        sum(1000 <= abs(elem) < 10_000 and abs(elem) % 10 == 6 for elem in seq[i: i + 3]) == 1,
        sum(seq[i: i + 3]) <= sub_max
    )):
        amount += 1
        max_sum = max(max_sum, sum(seq[i: i + 3]))

print(amount, max_sum)
# Answer: 507 1042
```

```python
with open("files/26.txt") as file:
    amount_of_notes = int(next(file))
    notes = [tuple(map(int, next(file).split())) for _ in range(amount_of_notes)]

notes = sorted(set(notes), key=lambda item: (item[0], item[1]))
students = {}
for note in notes:
    students[note[0]] = students.get(note[0], []) + [note[1]]

max_solved_task_in_row = answer_id = 0
for student_id, tasks in students.items():
    current_solved_task = max_solved_task = 1
    for i in range(1, len(tasks)):
        if tasks[i] - tasks[i - 1] == 2:
            current_solved_task += 1
        else:
            max_solved_task = max(max_solved_task, current_solved_task)
            current_solved_task = 1

    if max_solved_task > max_solved_task_in_row:
        max_solved_task_in_row = max_solved_task
        answer_id = student_id

print(answer_id, max_solved_task_in_row)
# Answer: 10135 42
```
